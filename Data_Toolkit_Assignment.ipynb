**Q2. How does broadcasting work in NumPy?**

Ans2- Broadcasting in NumPy is a powerful mechanism that allows operations between arrays of different shapes, or between an array and a scalar, without explicitly creating multiple copies of the smaller array to match the larger one. Instead, NumPy "virtually" stretches the smaller array to align with the larger one, enabling element-wise operations.

**Q13. Why is handling missing data important in Pandas?**

Ans13- Handling missing data in Pandas is crucial for several reasons, primarily to ensure the accuracy, reliability, and usability of data analysis and machine learning models.

Impact on Analysis and Models: Missing values, often represented as NaN (Not a Number) in Pandas, can lead to incorrect calculations, skewed statistics, and misleading conclusions if not addressed. Many analytical functions and machine learning algorithms cannot directly process NaN values and will either fail, produce errors, or yield unreliable results.

Maintaining Data Integrity: Unhandled missing data can compromise the integrity of the dataset. For instance, calculating averages or sums with missing values can lead to inaccurate representations of the data's central tendency or total, misrepresenting the underlying patterns.

Improved Model Performance: In machine learning, missing data can significantly degrade model performance. Addressing missing values through techniques like imputation (filling in missing values) or removal of rows/columns can help build more robust and accurate predictive models.

**Q17.  Explain the difference between apply() and map() in Pandas?**

Ans17- The primary difference between apply() and map() in Pandas lies in their scope of application and functionality:

map(): This method is exclusively for Pandas Series. It cannot be directly applied to an entire DataFrame. You would typically use map() on a single column (which is a Series) of a DataFrame.

apply(): This method is more versatile and can be used on both Pandas Series and Pandas DataFrames. When used on a DataFrame, it can apply a function along either rows (axis=0) or columns (axis=1).

```python
# Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame with numerical data
data = {
    'Feature_A': np.random.rand(50),
    'Feature_B': np.random.rand(50) * 2,
    'Feature_C': np.random.rand(50) + np.random.rand(50) * 0.5, # Some correlation
    'Feature_D': np.random.rand(50) * 3 - np.random.rand(50) # Some negative correlation
}
df_corr = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df_corr.corr()

print("Correlation Matrix:")
print(correlation_matrix)

# Visualize the correlation matrix using a Seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#Q8. Create a line plot with multiple lines using Matplotlib?
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for multiple lines
x = np.linspace(0, 10, 100) # 100 points between 0 and 10
y1 = np.sin(x) # First line: sine wave
y2 = np.cos(x) # Second line: cosine wave
y3 = x/5 # Third line: linear increase

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each line
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--')
plt.plot(x, y3, label='x/5', color='green', linestyle=':')

# Add labels, title, and legend
plt.title('Line Plot with Multiple Lines')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend() # Display the legend to differentiate lines
plt.grid(True) # Add a grid for better readability

# Show the plot
plt.show()
```

```python
#Q11. Perform matrix multiplication using NumPy?
import numpy as np

# Create two sample NumPy matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

# Perform matrix multiplication using the @ operator (preferred for clarity)
result_matrix_at = matrix1 @ matrix2

# Alternatively, using np.dot()
result_matrix_dot = np.dot(matrix1, matrix2)

print("\nMatrix multiplication result (using @ operator):")
print(result_matrix_at)

print("\nMatrix multiplication result (using np.dot()):")
print(result_matrix_dot)
```
